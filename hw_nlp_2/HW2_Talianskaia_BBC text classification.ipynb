# Data preparation.

As I failed to properly download already prepared data, I decied to take the raw one and to write it into csv-file with two columns: news texts and classes, using some tips from the web.

```python
# Writing CSV-file with news texts and classes
import os
import pandas as pd

data_folder = "C://For Python/bbc"
folders = ["business","entertainment","politics","sport","tech"]

os.chdir(data_folder)

x = []
y = []

for i in folders:
    files = os.listdir(i)
    for text_file in files:
        file_path = i + "/" +text_file
        print("reading file:", file_path)
        with open(file_path) as f:
            data = f.readlines()
        data = ' '.join(data)
        x.append(data)
        y.append(i)
   
data = {'news': x, 'type': y}       
df = pd.DataFrame(data)
print ('writing csv flie ...')
df.to_csv('../dataset.csv', index=False)
```

```python
#Converting text to vectorized representation. 
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')

features = tfidf.fit_transform(df.news).toarray()
labels = data.category_id
features.shape
```

I use TFiDF method instead of bag of words as for some topics where could be some common words and collactions (f.g. 'data analysis' for tech and business or 'economy' for politics and business). Thus, TFiDF approach takes this issue into account and helps to define 'the most specific' words and collacations.
Each of 2225 news texts is represented by 14454 features, representing the TFiDF score for different words and collacations.

Let's apply Chi-2 methode to our news texts to define the most correlated terms for each of topics.
Actually, it computes chi-squared stats between each non-negative feature and class.
All found words and collacations seem to be relevant for the topic.

```python
from sklearn.feature_selection import chi2
import numpy as np

N = 2
for Product, category_id in sorted(category_to_id.items()):
  features_chi2 = chi2(features, labels == category_id)
  indices = np.argsort(features_chi2[0])
  feature_names = np.array(tfidf.get_feature_names())[indices]
  unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
  bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
  print("# '{}':".format(Product))
  print("  . Most correlated words:\n       . {}".format('\n       . '.join(unigrams[-N:])))
  print("  . Most correlated collacatios:\n       . {}".format('\n       . '.join(bigrams[-N:])))
```
